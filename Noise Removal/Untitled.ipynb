```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

# Step 1: Load the Image
img = cv2.imread('noise1.png')  # Replace with your image path
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 2: Add Gaussian Noise
mean = 0
var = 0.01
sigma = var**0.5
gaussian = np.random.normal(mean, sigma, gray.shape)
noisy = np.clip(gray + gaussian * 255, 0, 255).astype(np.uint8)

# Step 3: Apply Gaussian Filter for Denoising
denoised = cv2.GaussianBlur(noisy, (5, 5), 1)

# Step 4: Evaluation Metrics
psnr_noisy = psnr(gray, noisy)
psnr_denoised = psnr(gray, denoised)
ssim_noisy = ssim(gray, noisy)
ssim_denoised = ssim(gray, denoised)

print(f"PSNR before denoising: {psnr_noisy:.2f} dB")
print(f"PSNR after denoising: {psnr_denoised:.2f} dB")
print(f"SSIM before denoising: {ssim_noisy:.3f}")
print(f"SSIM after denoising: {ssim_denoised:.3f}")

# Step 5: Show Images
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(noisy, cmap='gray')
plt.title("Noisy Image")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(denoised, cmap='gray')
plt.title("Denoised Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim

# Step 1: Load the noisy grayscale image
img = cv2.imread('noise1.png', cv2.IMREAD_GRAYSCALE)

# Check if image is loaded
if img is None:
    print("Error: Image not found!")
else:
    # Step 2: Apply various filters

    # 2.1 Median Filter (Recommended for salt & pepper)
    median_3 = cv2.medianBlur(img, 3)
    median_5 = cv2.medianBlur(img, 5)

    # 2.2 Gaussian Filter (smoothing residual noise)
    gaussian = cv2.GaussianBlur(median_5, (3, 3), 0.5)

    # 2.3 Bilateral Filter (preserves edges)
    bilateral = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)

    # 2.4 Hybrid: Median → Gaussian
    hybrid = cv2.GaussianBlur(median_5, (3, 3), 0.5)

    # Step 3: Display all results side-by-side
    titles = ['Original (Noisy)', 'Median (3x3)', 'Median (5x5)', 
              'Gaussian After Median', 'Bilateral Filter', 'Hybrid']
    images = [img, median_3, median_5, gaussian, bilateral, hybrid]

    plt.figure(figsize=(15, 8))
    for i in range(6):
        plt.subplot(2, 3, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

    # Step 4: Optional - If original image is available (for PSNR/SSIM)
    # Example dummy original (skip if you don't have ground truth)
    # original = cv2.imread('original_clean_image.png', cv2.IMREAD_GRAYSCALE)
    # print("PSNR Median 5x5:", psnr(original, median_5))
    # print("SSIM Median 5x5:", ssim(original, median_5))
```
